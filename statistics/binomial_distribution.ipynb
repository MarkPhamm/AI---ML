```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_coin_toss(n, num_trials):
    """
    Simulate coin toss.

    Parameters:
        n (int): Number of coin tosses.
        num_trials (int): Number of trials to run.

    Returns:
        numpy.ndarray: Array containing number of heads for each trial.
    """
    results = np.random.randint(2, size=(num_trials, n))  # 0 represents tails, 1 represents heads
    num_heads = np.sum(results, axis=1)
    return num_heads

def plot_distribution(num_heads, n):
    """
    Plot distribution of number of heads.

    Parameters:
        num_heads (numpy.ndarray): Array containing number of heads for each trial.
        n (int): Number of coin tosses.
    """
    plt.hist(num_heads, bins=np.arange(n + 2) - 0.5, rwidth=0.8, color='skyblue', edgecolor='black')
    plt.xticks(np.arange(n + 1))
    plt.xlabel('Number of Heads')
    plt.ylabel('Frequency')
    plt.title(f'Distribution of Heads in {n} Coin Tosses')
    plt.grid(axis='y', alpha=0.75)
    plt.show()

# Example usage:
n = 20  # Number of coin tosses
num_trials = 10000 # Number of trials to run

num_heads = simulate_coin_toss(n, num_trials)
plot_distribution(num_heads, n)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def plot_binomial_distribution(n, p):
    """
    Plot binomial distribution.

    Parameters:
        n (int): Number of trials.
        p (float): Probability of success in each trial.
    """
    x = np.arange(0, n + 1)
    pmf_values = binom.pmf(x, n, p)
    
    plt.bar(x, pmf_values, color='skyblue', edgecolor='black')
    plt.xlabel('Number of Heads')
    plt.ylabel('Probability')
    plt.title(f'Binomial Distribution (n={n}, p={p})')
    plt.grid(axis='y', alpha=0.75)
    plt.show()

# Example usage:
n = 40  # Number of trials (coin tosses)
p = 0.5 # Probability of success (getting heads)

plot_binomial_distribution(n, p)
```
